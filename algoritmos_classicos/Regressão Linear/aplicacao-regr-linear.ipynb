# Regressão Linear Simples e Múltipla

---



Desenvolveremos, neste notebook, algumas aplicações envolvendo a Regressão Linear (simples e múltipla). Como ferramentas de implementação, estaremos usando as bibliotecas `pandas`, `matplotlib` e a função `LinearRegression` da biblioteca `sklearn`. Como bases de dados, utilizaremos a relação entre a pressão e espessura de garrafas (Regressão Linear Simples) e a relação entre o tamanho, idade e preço de apartamentos (Regressão Linear Múltipla). Por fim, plotamos gráficos que representam os resultados obtidos após a realização das predições

## Aplicação de Regressão Linear Simples

---

Primeiro, vamos importar as bibliotecas `pandas` e `matplotlib`, para plotar os gráficos, separar e receber a base de dados

```
import pandas as pd
import matplotlib.pyplot as plt
```

Importaremos o dataset do GitHub com a função `read.csv` do `pandas` e armazenaremos em uma variável

```
base = pd.read_csv('https://raw.githubusercontent.com/Adriel-Cabral/regressao-linear/master/Garrafas.csv')
base
```

Plotar o gáfico de dispersão com a função `.scatter` para visualizarmos a correlação entre as variáveis

```
plt.scatter(base['pressao'], base['espessura'])  
plt.title('Correlação Linear') 
plt.xlabel('Pressão Aplicada') 
plt.ylabel('Espessura da Garrafa') 
```

Em seguida, vamos dividir nossas variáveis em duas categorias: dependente (espessura) e independente (pressão)

```
independente = base['pressao'].values
dependente = base['espessura'].values
independente
```

Precisamos transformar de um array 1d para 2d com a função `.reshape`, pois o modelo apenas recebe dessa forma

```
independente = independente.reshape(-1 ,1)
independente
```

A partir da biblioteca `sklearn`, vamos importar o modelo linear, criar o regressor, ajustar a melhor linha com base nos nossos dados e, por fim, predizer nossos resultados

```
from sklearn.linear_model import LinearRegression

regressor_simples =  LinearRegression()
regressor_simples.fit(independente, dependente)
resultados = regressor_simples.predict(independente)
```

Melhores coeficientes na regressão simples

```
print(regressor_simples.coef_, regressor_simples.intercept_)
```

Agora, vamos plotar nosso gráfico de dispersão: desenhamos a linha ajustada na cor vermelha, adicionamos o título "Regressão Linear Simples", colocamos na legenda do eixo x "Pressão Aplicada" e na legenda do eixo y "Espessura da Garrafa"

```
plt.scatter(independente, dependente)  
plt.plot(independente, resultados, color='red') 
plt.title('Regressão Linear Simples') 
plt.xlabel('Pressão Aplicada') 
plt.ylabel('Espessura da Garrafa') 
```

## Aplicação de Regressão Linear Múltipla

---



Importaremos um novo dataset do GitHub com a função do `pandas` e armazenaremos em uma variável

```
base2 = pd.read_csv('https://raw.githubusercontent.com/Adriel-Cabral/regressao-linear/master/Apartamento.csv')
base2
```

Vamos dividir nossas variáveis em duas categorias: dependente (preço) e independentes (idade do prédio e tamanho do apartamento)

```
independente2 = base2.iloc[:, 2:4].values
dependente2 = base2.iloc[:, 1].values
```

Nesse momento, vamos plotar nosso gráfico de correlação entre o preço e o tamanho: adicionamos o título "Gráfico de Correlação", colocamos na legenda do eixo x "Tamanho do Apartamento" e na legenda do eixo y "Preço do Apartamento"

```
plt.scatter(independente2[:, 0], dependente2)
plt.title('Gráfico de Correlação')
plt.xlabel('Tamanho do Apartamento')
plt.ylabel('Preço do Apartamento')
```

Também vamos plotar nosso gráfico de correlação entre o preço e a idade: adicionamos o título "Gráfico de Correlação", colocamos na legenda do eixo x "Idade do Prédio" e na legenda do eixo y "Preço do Apartamento"

```
plt.scatter(independente2[:, 1], dependente2)
plt.title('Gráfico de Correlação')
plt.xlabel('Idade do Prédio')
plt.ylabel('Preço do Apartamento')
```

A partir da biblioteca `sklearn`, vamos importar o modelo linear, criar o regressor, ajustar a melhor linha com base nos nossos dados e, por fim, predizer nossos resultados

```
#from sklearn.linear_model import LinearRegression

regressor_multiplo =  LinearRegression()
regressor_multiplo.fit(independente2, dependente2)
resultados = regressor_multiplo.predict(independente2)
resultados
```



Melhores coeficientes na regressão múltipla

```
print(regressor_multiplo.coef_, regressor_multiplo.intercept_)
```

Por fim, vamos plotar o gáfico de Regressão Linear Múltipla.
Para isso, precisamos importar a biblioteca numpy

```
# Importando a biblioteca numpy para criar o sistema de coordendas
import numpy as np

# Criando nosso sistema de coordenadas x e y
x, y = np.meshgrid(independente2[:, 0],independente2[:, 1])

# Criando a coordenada z
z = x * regressor_multiplo.coef_[0] + y * regressor_multiplo.coef_[1] +  regressor_multiplo.intercept_

# Escolhendo o tamanho do grafico/criando grafico
fig = plt.figure(figsize=(7, 7))

# Adicionando o plano ao grafico
ax = fig.add_subplot(projection = '3d')

# Plotando o gráfico com temperatura (para visualizar o aumento do preço)
ax.plot_surface(x,y,z, cmap='viridis')

plt.title('Gráfico de Regressão Múltipla')
ax.set_xlabel('Tamanho do Apartamento')
ax.set_ylabel('Idade do Prédio')
ax.set_zlabel('Preço do Apartamento')
```
